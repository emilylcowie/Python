```python
import matplotlib.pyplot as plt
import pandas as pd

path = '/content/Large_Data_Set.xlsx'
df = pd.read_excel(path, sheet_name='Beijing May-Oct 1987', skiprows=5)
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Daily Mean Air Temperature'], marker='o')
plt.title('Temp by date')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Convert the 'Date' column to datetime, force bad values to NaT
df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)

# Drop rows where the date couldn't be parsed (NaT)
df = df.dropna(subset=['Date'])

# Plot as before
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Daily Mean Air Temperature'], marker='o')
plt.title('Daily Mean Air Temperature by Date Beijing 1987')
plt.xlabel('Date')
plt.ylabel('Temp (degrees celcius)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

mean_temp = df['Daily Mean Air Temperature'].mean()
print("Mean Temp:", mean_temp)

```

```python
mean_temp = df['Daily Mean Air Temperature'].mean()
print("Mean Temp:", mean_temp)

min_temp = df['Daily Mean Air Temperature'].min()
max_temp = df['Daily Mean Air Temperature'].max()
temp_range = max_temp - min_temp

print("Minimum Temp:", min_temp)
print("Maximum Temp:", max_temp)
print("Range of Temp:", temp_range)
```

```python
import matplotlib.dates as mdates

import warnings
warnings.filterwarnings("ignore", message="Print area cannot be set to Defined name")

df2 = pd.read_excel(path, sheet_name='Beijing May-Oct 2015', skiprows=5)

df['Date'] = pd.to_datetime(df['Date'], errors='coerce', dayfirst=True)
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce', dayfirst=True)
df['MonthDay'] = df['Date'].apply(lambda x: x.replace(year=2000))
df2['MonthDay'] = df2['Date'].apply(lambda x: x.replace(year=2000))


plt.figure(figsize=(10, 6))

plt.plot(df['MonthDay'], df['Daily Mean Air Temperature'], marker='o', label='Temperature 1987')
plt.plot(df2['MonthDay'], df2['Daily Mean Air Temperature'], marker='s', label='Temperature 2015')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())


# Labels and formatting
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temp')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
window_size = 5  # Adjust for more or less smoothing

df['Temp_smooth'] = df['Daily Mean Air Temperature'].rolling(window=window_size, center=True).mean()
df2['Temp2_smooth'] = df2['Daily Mean Air Temperature'].rolling(window=window_size, center=True).mean()

plt.plot(df['MonthDay'], df['Temp_smooth'], label= 'Temperature 1987')
plt.plot(df2['MonthDay'], df2['Temp2_smooth'], label='Temperature 2015')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
```
